**Spark Declarative Pipeline (SDP)** (formerly DLT)

````sql
create schema if not exists sony_dev.bronze;
create schema if not exists sony_dev.silver;
create schema if not exists sony_dev.gold;
````

````sql
-- Test: Verify that bronze, silver, and gold schemas exist in sony_dev
SELECT schema_name,
       CASE WHEN schema_name IN ('bronze', 'silver', 'gold') THEN 'PASS' ELSE 'FAIL' END AS test_result
FROM   sony_dev.information_schema.schemata
WHERE  schema_name IN ('bronze', 'silver', 'gold')
ORDER BY schema_name;
````

````sql
-- Test: Assert exactly 3 schemas (bronze, silver, gold) exist
SELECT
  CASE
    WHEN COUNT(*) = 3 THEN 'ALL TESTS PASSED: 3/3 schemas found'
    ELSE CONCAT('FAIL: Expected 3 schemas, found ', COUNT(*))
  END AS test_summary
FROM sony_dev.information_schema.schemata
WHERE schema_name IN ('bronze', 'silver', 'gold');
````
